```python
import pandas as pd
```

The goal for this code is to create a dataframe from an xls file from the brazilian infrastructure department, which holds information from hydroeletrical powerplant water levels across the country. In order to work with nested loops, dictionary format will be easier to handle.

```python
sheets_dict = pd.read_excel('ACOMPH_12.05.2020.xls', sheet_name=None)
```

```python
all_sheets = []


df_full_complete = pd.DataFrame()


# Loop nas guias (regiao)
for regiao, sheet in sheets_dict.items():

    # Pegando os nomes dos locais nas guias
    locais = sheet.iloc[0,1:].dropna()
    
    #tamanho do intervalo de cada local
    BLOCK_SIZE= 8

    df_full_region = pd.DataFrame()
    
    # O loop interno que vai nas regiões(abas) e retira os dados dos locais
    for i, local in enumerate(locais):

        # BASE é o primeiro indice da coluna do bloco em analise...
        base = 1 + BLOCK_SIZE * i
        col_start, col_end = base, base + 8
        df_aux = sheet.iloc[4:34, col_start:col_end]

        
        col_names = ["Lido", "Consol", "Lido", "Consol", "Lido", "Consol", "icr_consol", "nat_consol"]

        df_aux.columns = col_names

        df_aux.reset_index(inplace=True, drop=True)

        df_aux['local'] = local
        df_aux['regiao'] = regiao
        df_aux['data'] = sheet.iloc[4:34,0].reset_index(drop=True)
        
        df_full_region = pd.concat([df_full_region, df_aux], axis=0)
    
    
    # Saindo do loop da regiao, ele vai entrar, ele insere o bloco no full
    
    df_full_complete = pd.concat([df_full_complete, df_full_region], axis=0)
```
